# Battery Thermal Analysis Jupyter MDK Demo

This notebook demonstrates a second in-tool MDK scenario: a Jupyter thermal analysis updates MMS with a requirement, constraint, and test case.

```python
import sys
from pathlib import Path

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if (candidate / "mdk").exists() and (candidate / "sysml_docgen").exists():
        if str(candidate) not in sys.path:
            sys.path.insert(0, str(candidate))
        print(f"Using project root: {candidate}")
        break

%load_ext mdk.jupyter.sysml_docgen_notebook
%sysml_config --server http://127.0.0.1:8000 --project satellite-power --branch main --user engineer
%sysml_health
```

```python
%%sysml_requirement REQ-JUP-020 "Battery thermal limit requirement" --owner "Thermal Analysis Team" --satisfy BLK-BATTERY
Battery maximum temperature shall not exceed 45 degC during peak discharge.
```

```python
from mdk.jupyter.sysml_docgen_notebook import result_constraint

ambient_temp = 22.0
thermal_rise = 18.6
battery_temp_max = ambient_temp + thermal_rise

result_constraint(
    "CST-JUP-020",
    "Battery temperature constraint",
    expression=f"T_battery_max = {battery_temp_max:.1f} degC <= 45 degC",
    owner="Thermal Analysis Team",
    constrains="REQ-JUP-020",
)
```

```python
from mdk.jupyter.sysml_docgen_notebook import test_case

assert battery_temp_max <= 45.0

test_case(
    "TST-JUP-020",
    "Battery thermal simulation",
    method=f"Jupyter thermal simulation: max temperature = {battery_temp_max:.1f} degC",
    criterion="T_battery_max <= 45 degC",
    owner="Thermal Analysis Team",
    verifies="REQ-JUP-020",
)
```
